```python
# 0.필요모듈 import
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

import pandas as pd
```

```python
# 1. 데이터셋 로드 (보스턴 집값 데이터)
boston_df = pd.read_csv('DataSet/boston.csv')
boston_df

# 1.1 변수 타입 및 속성 확인
# print(type(boston_df))
# print(dir(boston_df))

```

```python
# 2. DataFrame 데이터를  data와 target으로 분리
# data
X = boston_df.drop(columns='MEDV')
print("data : \n {}".format(X))
# target
y = boston_df[['MEDV']]
print("target : \n {}".format(y))
```

```python
# 3. 다항속성 생성
# 2차 다항회귀 설정
polynomial_transformer = PolynomialFeatures(2)

# 2차 속성 생성
polynomial_data = polynomial_transformer.fit_transform(X)
print("polynomial_data : \n {}".format(polynomial_data))

# 2차 속성의 이름
polynomial_feature_names = polynomial_transformer.get_feature_names_out(X.columns)
print("polynomial_feature_names : \n {}".format(polynomial_feature_names))
```

```python
# 4. ndarray 형태로 산출된 다항 속성과 이름을  pd.DataFrame으로 변환
X = pd.DataFrame(polynomial_data, columns=polynomial_feature_names)
print("polynomial data : \n {}".format(X))
y = boston_df[['MEDV']]
print("target : \n {}".format(y))
```

```python
# 6. 머신러닝 알고리즘 사용을 위한 train, test set 구성
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=5)
```

```python
# 7. 머신러닝을 위한 모델 선정 및 학습
# LinearRegression 사용
model = LinearRegression()
model.fit(X_train, y_train)
print(type(model))
print(dir(model))
```

```python
# 최적화된 theta값 확인
print("theta 0 : \n", model.intercept_)
print("theta 1 ~ n  : \n",model.coef_)
```

```python
# 8. 최적화된 모델을 바탕으로 Test data 예측
y_test_prediction = model.predict(X_test)

# Test datd와 예측값의 오차를 RMSE 로 나타냄
RMSE = mean_squared_error(y_test, y_test_prediction) ** 0.5
print("RMSE is : {}".format(RMSE))
```
