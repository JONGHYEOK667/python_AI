# CelebA 데이터셋의 주요 특징

 1. 데이터셋 구성 : 약 20만 개 이상의 이미지로 구성된 대규모 얼굴 데이터셋.
 
 2. 이미지 크기 : 178 × 218 하지만 필요에 따라 크기를 조정하여 사용.
 
 3. 얼굴 속성(Attributes) : 각 이미지에는 40개의 얼굴 속성이 레이블로 제공됩니다.
    예 : Smiling, Male, Eyeglasses, Young, Wearing Hat 등.
    
 4. 얼굴 좌표(Facial Landmarks) : 눈, 코, 입 등 얼굴의 주요 부분에 대한 좌표 정보(5개 랜드마크 제공).
 
 5. 이미지 다양성 : 다양한 배경, 조명 조건, 얼굴 각도 및 표정을 포함.
 
 6. 사용 라이선스 : 연구 및 비상업적 목적으로 무료로 사용 가능.

# STEP 0. 필요 모듈 임포트

```python
# 0. 필요한 모듈 임포트

import os
import tensorflow as tf
import tensorflow_datasets as tfds
from tensorflow import keras
from keras.models import Model
from keras.layers import Conv2D, MaxPool2D, Input, Flatten, Dense, Concatenate
from keras.models import Sequential
from keras.utils import to_categorical

import imageio
from skimage.transform import resize

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from IPython.display import Image

warnings.filterwarnings('ignore')
%matplotlib inline

SEED = 34
```

# STEP 1. 도입전 기본처리



```python
# celeb_a 데이터를 tensorflow-dataset에서 다운받는게 쉽지않음
# 데이터를 수동으로받아 처리해보려했는데 실패함
# 미리 제공된 celeba_small데이터를 이용해 진행

# train_images, test_images, train_labels, test_labels를 np.array로 로드
celeba_small = np.load('DataSet/celeba_small.npz')
train_images = celeba_small['train_images']
train_labels = celeba_small['train_labels']

test_images = celeba_small['test_images']
test_labels = celeba_small['test_labels']
```

# STEP 2. 데이터 전처리

```python
# 이미지 데디터 범위 확인
print(train_images.shape, test_images.shape)
print(train_images.dtype, test_images.dtype)
```

```python
# images/lables 이미지 데이터의 최소/최대값을 출력
print(train_images.min(), test_images.min())
print(train_images.max(), test_images.max())
print(train_labels.min(), test_labels.min())
print(train_labels.max(), test_labels.max())
```

```python
# labels의 shape을 (배치,2)에서 (배치,2),(배치,2)로 변경 (배치,(성별구분,웃음구분)) -> (배치,남,여),(배치,웃음,안웃음)
train_male_labels, train_smile_labels = np.split(train_labels, 2, axis=1)
test_male_labels, test_smile_labels = np.split(test_labels, 2, axis=1)
random_int = np.random.randint(0, len(train_labels))
print(train_male_labels.shape, train_smile_labels.shape)
print(train_male_labels[random_int], train_smile_labels[random_int], train_labels[random_int])

train_male_labels = to_categorical(train_male_labels)
train_smile_labels = to_categorical(train_smile_labels)
test_male_labels = to_categorical(test_male_labels)
test_smile_labels = to_categorical(test_smile_labels)
print(train_male_labels.shape, train_smile_labels.shape)
print(train_male_labels[random_int], train_smile_labels[random_int], train_labels[random_int])
```

```python
# lable의 shape을 (배치,2),(배치,2)에서 (배치,4)로 변경
# (배치,남,여),(배치,웃음,안웃음) -> (배치, 남,여,웃음,안웃음)
train_labels2 = np.concatenate([train_male_labels, train_smile_labels], axis=1)
test_labels2 = np.concatenate([test_male_labels, test_smile_labels], axis=1)
print(train_labels2.shape, test_labels2.shape)
```

# STEP 4. 모델링

```python
# smile과 gender 학습셋을 각각 처리하는 이미지 classification 모델 2개 만들기 
def simple_model():
    inputs = Input((72,59,3))
    
    x = Conv2D(32, 3, activation='relu')(inputs)
    x = MaxPool2D(2)(x)
    x = Conv2D(64, 3, activation='relu')(x)
    x = MaxPool2D(2)(x)
    x = Conv2D(64, 3, activation='relu')(x)
    x = MaxPool2D(2)(x)
    
    x = Flatten()(x)
    x = Dense(64, activation='relu')(x)
    
    outputs = Dense(2,activation='softmax')(x)
    model = Model(inputs,outputs)
    
    return model
```

```python
# 2개의 모델을 할당받고 서머리 출력
gender_model = simple_model()
smile_model = simple_model()

gender_model.summary()
```

```python
smile_model.summary()
```

```python
# 만든 모델에 로스와 옵티마이저, 메트릭 설정
gender_model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
smile_model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
gender_model.get_weights()[0][0][0][0]
```

```python
smile_model.get_weights()[0][0][0][0]
```

```python
# 만든 모델에 train_images를 학습
gender_hist = gender_model.fit(train_images, train_male_labels, validation_data=(test_images, test_male_labels), epochs=15, verbose=1)
smile_hist = smile_model.fit(train_images, train_smile_labels, validation_data=(test_images, test_smile_labels), epochs=15, verbose=1)
```

```python
# 학습 진행 사항을 plt 로 출력
plt.plot(gender_hist.history['accuracy'], label = 'gender_accuracy')
plt.plot(gender_hist.history['loss'], label = 'gender_loss')
plt.plot(gender_hist.history['val_accuracy'], label = 'gender_val_accuracy')
plt.plot(gender_hist.history['val_loss'], label = 'gender_val_loss')

plt.plot(smile_hist.history['accuracy'], label = 'smile_accuracy')
plt.plot(smile_hist.history['loss'], label = 'smile_loss')
plt.plot(smile_hist.history['val_accuracy'], label = 'smile_val_accuracy')
plt.plot(smile_hist.history['val_loss'], label = 'smile_val_loss')

plt.legend(loc='upper left')
plt.show()
```

# STEP 6. 멀티아웃풋 모델링

```python
# 해당 학습셋을 처리하는 이미지 classification 모델을 작성 (단 레이블2개 한번에 처리)
def multi_model():
    inputs = Input((72,59,3))
    
    x = Conv2D(32, 3, activation='relu')(inputs)
    x = MaxPool2D(2)(x)
    x = Conv2D(64, 3, activation='relu')(x)
    x = MaxPool2D(2)(x)
    x = Conv2D(64, 3, activation='relu')(x)
    x = MaxPool2D(2)(x)
    
    # 각 이미지에서 성별, 웃음여부를 판단하기 위해 각 얼굴로 부터 특징점을 찾는 과정은 상당히 비슷학 것으로 생각 
    # (특징점을 어떻게 해석할지는 다르더라도)
    # 따라서 사람얼굴에서 특징 (latent_vector)을 추출하는 부분은 통일하고 해석부분을 다르게 가져갈때 
    # 모델의 사이즈를 줄이는데 도움이 된다.
    
    
    x = Flatten()(x)
    latent_vector = Dense(64, activation='relu')(x)
    
    y1 = Dense(32,activation='relu')(latent_vector)
    y1 = Dense(16,activation='relu')(y1)
    y1 = Dense(8,activation='relu')(y1)
    gender_outputs = Dense(2,activation='softmax')(y1)
    
    y2 = Dense(32,activation='relu')(latent_vector)
    y2 = Dense(16,activation='relu')(y2)
    y2 = Dense(8,activation='relu')(y2)
    smile_outputs = Dense(2,activation='softmax')(y2)
        
    model = Model(inputs, [gender_outputs, smile_outputs])
    return model
```

```python
# 모델을 할당받고 서머리를 출력
model = multi_model()
model.summary()
```

```python
# 만들 모델에 로스와 옵티마이저, 메트릭 설정
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# 만든 모델에 train_images 학습
model_hist = model.fit(train_images, [train_male_labels, train_smile_labels], validation_data=(test_images, [test_male_labels, test_smile_labels]), epochs=15, verbose=1)
```

```python
plt.plot(model_hist.history['dense_8_accuracy'], label = 'gender_accuracy')
plt.plot(model_hist.history['dense_8_loss'], label = 'gender_loss')
plt.plot(model_hist.history['val_dense_8_accuracy'], label = 'gender_val_accuracy')
plt.plot(model_hist.history['val_dense_8_loss'], label = 'gender_val_loss')

plt.plot(model_hist.history['dense_12_accuracy'], label = 'gender_accuracy')
plt.plot(model_hist.history['dense_12_loss'], label = 'gender_loss')
plt.plot(model_hist.history['val_dense_12_accuracy'], label = 'gender_val_accuracy')
plt.plot(model_hist.history['val_dense_12_loss'], label = 'gender_val_loss')

plt.legend(loc='upper left')
plt.show()
```

```python
# 뒷단을 수정하여 smile 모델을 분리하시오
smile_model = Model(inputs = model.input, outputs = model.get_layer('dense_12').output)
smile_model.summary()
```
