```python
# K-means 장점
# 변수들에 대한 배경지식, 역할, 영향도에 대해 모르더라도 데이터사이의 거리만 구할 수 있으면 사용가능
# 알고리즘이 쉬운 수식으로 되어있어 이해와 해석이 용이함
# --> 어떠한 데이터에도 적용하기 쉬우며 모델에 대한 이해와 해석이 간단


# k-means 단점
# 최적의 클러스터 개수 k 정하는것이 어려움 (Elbow methed는 보조수단일 뿐 정답이 아님)
# 이상치의 영향을 많이 받음 --> centorid업데이트 과정에서 영향을 크게 미침
# centroid 의 초기 위치에의해 연산속도 및 수렴여부 영향을 끼침   --> k-means++ 를 통해 개선할 수 있음 (파라미터 init='k-means++')
# 차원이 놓은 데이터에 적용할 때 성능 저하
```

```python
# 0.필요모듈 import
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans

```

```python
# 1. 데이터셋 로드 (csv -> dataframe)
# 판다스 출력 결과를 소수점 둘째 자리 까지로 제한
pd.options.display.float_format = '{:,.2f}'.format

sales_df = pd.read_csv('DataSet/sales_data.csv', index_col=['customer_id'])
sales_df
```

```python
# 2. 데이터셋 시각화  (클러스터링 전)
sns.set(style='darkgrid',
        rc={'figure.figsize':(16, 9)})

# 데이터 시각화
sns.scatterplot(x=sales_df['total_buy_cnt'], y=sales_df['total_price'], s=200)
```

```python
# 3. 제이터 이상치 제거 
# IQR 방법 사용
def get_outlier_mask(df, weight=1.5):   
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)

    IQR = Q3 - Q1
    IQR_weight = IQR * weight

    range_min = Q1 - IQR_weight
    range_max = Q3 + IQR_weight

    outlier_per_column = (df < range_min) | (df > range_max)
    
    is_outlier = outlier_per_column.any(axis=1)

    return is_outlier

outlier_idx_cust_df = get_outlier_mask(sales_df, weight=1.5)

# 아웃라이어 제거한 데이터 프레임만 추가
sales_df = sales_df[~outlier_idx_cust_df]

# 아웃라이어 제거한 데이터프레임 시각화
sns.scatterplot(x=sales_df['total_buy_cnt'], y=sales_df['total_price'], s=200)
```

```python
# 4. 데이터 표준화 (Stanardization)
df_mean = sales_df.mean()  # 각 컬럼의 평균값
df_std = sales_df.std()  # 각 컬럼의 표준편차

scaled_df = (sales_df - df_mean)/df_std  # 컬럼별 표준화 진행 

scaled_df.columns = ['total_buy_cnt', 'total_price']


# 인덱스 설정
scaled_df.index = sales_df.index 

# 아웃라이어 제거한 표준화 데이터프레임 시각화
sns.scatterplot(x=scaled_df['total_buy_cnt'], y=scaled_df['total_price'], s=200)
scaled_df
```

```python
# 5. 클러스터링 모델 선언 및 학습 (k-means)
# Elbow method 기반 최적 클러스터 갯수 찾기
# inertia (각 centroid 와 데이터 사이의 거리제곱의 총 합)값 저장을위한 리스트 선언
inertias = []

for k in range(1, 16):
    model = KMeans(n_clusters=k, random_state=123)
    model.fit(scaled_df)
    inertias.append(model.inertia_)
    
# k값에 따른 inertia값 시각화
sns.lineplot(x=range(1, 16), y=inertias, marker='o')
```

```python
# 6. 클러스터링 모델 선언 및 학습 (k-means)
model = KMeans(n_clusters=5, random_state=123)
model.fit(scaled_df)

# inertia (각 centroid 와 데이터 사이의 거리제곱의 총 합)값 확인
print(model.inertia_)
```

```python
# 7. 클러스터 시각화 --> 스케일 데이터는 클러스터링까지만, 이후부터는 데이터 원본으로
# label 컬럼 생성
sales_df['label'] = model.predict(scaled_df)

sns.scatterplot(x=sales_df['total_buy_cnt'], y=sales_df['total_price'], hue=sales_df['label'], s=200, palette='bright')
```
